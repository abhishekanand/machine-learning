```python
# Load libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Without Seaborn  
```

```python
# Load the iris data set from remote url
def loadDatasetFromUrl():
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
    #giturl = "https://raw.githubusercontent.com/abhishekanand/datasets/master/data/iris.data" # Use raw URL
    columnattribute = ['sepal-length (cm)', 'sepal-width (cm)', 'petal-length (cm)', 'petal-width (cm)', 'species']
    return pd.read_csv(url, names=columnattribute)
    #return pd.read_csv(giturl, names=columnattribute)
```

```python
iris_dataset = loadDatasetFromUrl()
iris_dataset.head() # #peek at the data 
```

# Data Cleaning

```python
iris_dataset['species'] = iris_dataset['species'].str.replace('Iris-','')
iris_dataset.head()
```

# Vizual Exploration

```python
iris_dataset.plot(kind='line', subplots=True, layout=(2, 2), sharex=False, sharey=False, figsize= (10,10) )
plt.show()
```

```python
# Univariate Plots : box and whisker plots of each 
iris_dataset.plot(kind='box', subplots=True, layout=(2, 2), sharex=False, sharey=False, figsize= (10,10))
plt.show()
```

```python
# Univariate Plots : Histogram plots of each 
iris_dataset.plot(kind='hist', subplots=True, layout=(2, 2), sharex=False, sharey=False, figsize= (10,10))
plt.show()
```

```python
# Univariate Plots : Histogram plots of each 
iris_dataset.hist(figsize=(10, 10), bins=6, xlabelsize=8, ylabelsize=8)
plt.show()
```

```python
# Univariate Density Plots
iris_dataset.plot(kind= 'density',figsize=(10, 10))
#iris_dataset.plot(kind='density', subplots=True, layout=(3,3),figsize=(10, 10), sharex=False)
plt.show()
```

```python
# multivariate plot to spot structured relationships between input variables
from pandas.plotting import scatter_matrix
scatter_matrix(iris_dataset, figsize= (15,15), marker='o', alpha=.8)
plt.show()
```
