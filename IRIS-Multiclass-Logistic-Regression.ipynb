```python
# Load libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Load the iris data set from remote url
def loadDatasetFromUrl():
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
    #giturl = "https://raw.githubusercontent.com/abhishekanand/datasets/master/data/iris.data" # Use raw URL
    columnattribute = ['sepal-length (cm)', 'sepal-width (cm)', 'petal-length (cm)', 'petal-width (cm)', 'species']
    return pd.read_csv(url, names=columnattribute)
    #return pd.read_csv(giturl, names=columnattribute)
```

```python
iris_dataset = loadDatasetFromUrl()
iris_dataset.head() # #peek at the data 
```

```python
iris_dataset['species'] = iris_dataset['species'].str.replace('Iris-','')
iris_dataset.head()
```

```python
from sklearn import model_selection
```

```python
# Split-out validation dataset
array = iris_dataset.values
X = array[:,0:4]
Y = array[:,4]
validation_size = 0.20
seed = 7
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, Y, test_size=validation_size, random_state=seed)
```

```python
# from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
```

```python
#creating Model
logreg = LogisticRegression()
```

```python
# Train Model
logreg.fit(X_train, Y_train)
```

```python
#Predicting the test set results 
Y_pred = logreg.predict(X_validation)
```

```python
result = pd.DataFrame( {'X_validation':list( X_validation),'Y_validation': Y_validation, 'Y_pred': Y_pred}, columns=['X_validation','Y_validation', 'Y_pred'])
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y_validation, Y_pred)
print(confusion_matrix)
```
